```python
import numpy as np
```

```python
def jacobi(u, eps=1.0):
    x = u.copy()
    y = u.copy()
    
    delt = eps + 1.0
    
    while delt > eps:
        for i in range(1, u.shape[0]-1):
            for j in range(1, u.shape[1]-1):
                y[i,j] = (x[i-1,j] + x[i+1,j] + x[i,j-1] + x[i,j+1]) / 4.0
        delt = np.sum( np.abs(x - y) )
        x, y = y, x
        
    return x
```

```python
u = np.zeros((10,10))
```

```python
u[0,:]=100.0
```

```python
def gaussian(a,b,c,x):
    return a * np.exp(-(x-b)**2 / 2*c**2)
```

```python
samp = np.random.randn(50000)
```

```python
plt.plot(np.arange(-4,4,0.1), gaussian(1/np.sqrt(2*np.pi),0,1,np.arange(-4,4,0.1)))
plt.show()
```
